````python
from transformers import pipeline
from langchain_core.prompts import PromptTemplate
````

````python
# Load free GPT2 model
generator = pipeline("text-generation", model="gpt2")
````

````python
response = generator(
    "What is Artificial Intelligence?",
    max_new_tokens=50
)

print(response[0]['generated_text'])
````

````python
prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

text = prompt.format(topic="Machine Learning")

response = generator(text, max_length=60)
print(response[0]['generated_text'])
````

````python
def chain(topic):
    text = prompt.format(topic=topic)
    return generator(text, max_length=60)[0]['generated_text']

print(chain("Deep Learning"))
````

````python
topics = ["AI", "NLP", "Computer Vision"]

for t in topics:
    print("\nTopic:", t)
    print(chain(t))
````

## 1. Introduction to LangChain

### 🔹 What is LangChain?

LangChain is a framework designed to build applications powered by Large Language Models (LLMs). Instead of directly calling an LLM, LangChain provides structured components that help in building scalable and modular AI systems.

It allows developers to combine prompts, models, tools, and memory into a single workflow.

---

### 🔹 Why is it important in modern LLM ecosystem?

Modern AI applications require more than just text generation. They need:

* context awareness
* interaction with external tools
* multi-step reasoning

LangChain solves this by providing a structured way to connect all these components.

---

### 🔹 Problems it solves

LangChain helps solve the following challenges:

* **Orchestration** → Managing multiple steps in LLM workflows
* **Chaining** → Connecting outputs of one step to another
* **Tool Integration** → Using APIs, calculators, databases
* **Memory Handling** → Maintaining conversation history

---

## 2. Core Components of LangChain

---

### 🔹 1. LLMs and Chat Models

**Concept:**
LLMs are models that generate human-like text based on input.

**Why it exists:**
They act as the core engine of any GenAI application.

**Code Example:**

```python id="llmex"
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

response = generator("What is AI?", max_new_tokens=50)
print(response[0]['generated_text'])
```

---

### 🔹 2. Prompts and Prompt Templates

**Concept:**
PromptTemplates allow dynamic input instead of hardcoded prompts.

**Why it exists:**
Improves reusability and flexibility.

**Code Example:**

```python id="promptex"
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

print(prompt.format(topic="Machine Learning"))
```

---

### 🔹 3. Chains

**Concept:**
Chains connect multiple operations together.

**Why it exists:**
To create multi-step workflows instead of single LLM calls.

**Code Example:**

```python id="chainex"
def chain(topic):
    text = prompt.format(topic=topic)
    return generator(text, max_new_tokens=60)[0]['generated_text']

print(chain("Deep Learning"))
```

---

### 🔹 4. Memory

**Concept:**
Memory stores past interactions.

**Why it exists:**
To maintain conversation context.

**Example Explanation:**
A chatbot remembers previous questions to give better responses.

---

### 🔹 5. Agents

**Concept:**
Agents decide which tool to use based on input.

**Why it exists:**
To enable decision-making and automation.

---

### 🔹 6. Tools

**Concept:**
External functions or APIs that LLM can use.

**Why it exists:**
LLMs alone cannot perform real calculations or fetch real data.

---

### 🔹 7. Document Loaders

**Concept:**
Load external data like text files or PDFs.

**Why it exists:**
To use real-world data in LLM applications.

**Code Example:**

```python id="loaderex"
from langchain_community.document_loaders import TextLoader

loader = TextLoader("sample.txt")
docs = loader.load()
print(docs)
```

---

## 3. Architecture Explanation

### 🔹 Flow of LangChain System

```text id="flowex"
User Input → Prompt → LLM → Chain → Tool/Agent → Output
```

---

### 🔹 Explanation of Flow

1. **User Input** → User provides query
2. **Prompt** → Input is structured using PromptTemplate
3. **LLM** → Model processes the input
4. **Chain** → Multiple steps are connected
5. **Tool/Agent** → External tools are used if required
6. **Output** → Final response is generated
